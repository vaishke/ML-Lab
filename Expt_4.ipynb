## Find-S Algorithm

```python
dataset = [['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
           ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
           ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
           ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']]


# S_0 = ["", "", "", "", "", ""]
# S_1 = [Sunny, Warm, Normal, Strong, Warm, Same]
# S_2 = [Sunny, Warm, ?, Strong, Warm, Same]
# S_3 = [?, ?, ?, Strong, Warm, ?]
# s_4 = [?, ?, ?, Strong, ?, ?]
```

```python
def generalize(record: list[str], hypothesis: list[str]) -> list[str]:
    for i in range(len(hypothesis)):
        if hypothesis[i] == "?":
            continue
        if(hypothesis[i] == ""):
            hypothesis[i] = record[i]
        if hypothesis[i] != record[i]:
            hypothesis[i] = "?"
    return hypothesis

def find_S(dataset: list[list[str]]) -> list[str]:
    hypothesis = [""] * len(dataset[0])
    for record in dataset:
        target = record[-1]
        if target == 'Yes':
            hypothesis = generalize(record, hypothesis)
        else:
            continue
    return hypothesis[:-1]


print(find_S(dataset))
```
