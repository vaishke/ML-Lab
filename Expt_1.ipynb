```python
def mean(data: list[int]) -> float:
    return sum(data) / len(data)

def median(data: list[int]) -> int:
    data.sort()
    if(len(data) % 2 != 0):
        return data[len(data) // 2]
    else:
        return (data[len(data) // 2 - 1] + data[len(data) // 2]) // 2

def mode(data: list[int]) -> int:
    freq = {}
    for num in data:
        freq[num] = freq.get(num, 0) + 1
    max_freqs = [key for key, value in freq.items() if value == max(freq.values())]
    return min(max_freqs)

def variance(data: list[int]) -> float:
    meann = mean(data)
    total = 0
    for x in data:
        total += (x - meann) ** 2
    return total / len(data)

def std_deviation(data: list[int]) -> float:
    return variance(data) ** 0.5

data = [48, 50, 52, 50, 49, 51, 50, 53, 47, 50]
print("Data: ", data)
print("Mean = ", mean(data))
print("Median = ", median(data))
print("Mode = ", mode(data))
print("Variance = ", variance(data))
print("Standard Deviation = ", std_deviation(data))
```
